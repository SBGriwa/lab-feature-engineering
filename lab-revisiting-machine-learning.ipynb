```python
#geocode2 is a cat 
# wealth1 : lot of nulls so bfill method for more balance
# adi, dma, msa not so much NaNs and biased : median
```

```python
def log_transform(x):
    if np.isfinite(x) and x!=0:
        return np.log(x)
    else:
        return np.NAN
```

```python
data['adi'] = data['adi'].apply(log_transform).fillna(data['adi'].median())
data['dma'] = data['dma'].apply(log_transform).fillna(data['dma'].median())
data['msa'] = data['msa'].apply(log_transform).fillna(data['msa'].median())
data['wealth1'] = data['wealth1'].apply(log_transform).fillna(method='bfill')
```
